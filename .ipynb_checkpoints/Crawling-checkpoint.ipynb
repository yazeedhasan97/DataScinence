```python
def scrape_data(url):
    data = requests.get(url) # 
    
    try:
        data.raise_for_status()
    except Exception as e:
        raise e
        
    data = data.text
    
    parser = BeautifulSoup(data, 'html.parser')
    links = [
        i['href'].strip() for i in obj.find_all('a', href=True) if i['href'].split('.')[-1].strip() in  ALLOWED_EXTENTIONS
    ]
    
    return links

def get_file(url, path=None, stream=True, chunk_size=4096):
    name = url.split('/')[-1]
    if path is None:
        path = os.path.join(DATA_DIR, name)
        
    try:
        data = requests.get(url, stream=stream)
        data.raise_for_status()
   
        with open(path, 'bw') as file:
            for chunk in data.iter_content(chunk_size=chunk_size):
                if chunk:
                    file.write(chunk)

        # record_from_db.downloaded = True
        # session.commit()

    # Error Spacific Catching # The order in which you catch the errors is important 
    except requests.exceptions.HTTPError as e:
        raise e
    except requests.exceptions.ConnectionError as e:
        raise e
    except requests.exceptions.Timeout as e:
        raise e
    except requests.exceptions.RequestException as e:
        raise e
    except FileNotFoundError as e:
        raise e
    except IOError as e:
        raise e
    except Exception as e:
        raise e

    return path


def filter_files(files: list, dates: list[str]=None, ttype: TaxiTypes = None,):

    def type_filter(file, ):
        name = file.split('/')[-1]
        typ = name.split('_')[0]
        # print(typ)
        # print(typ == ttype)
        return typ == ttype.value

    def date_filter(file, ):
        name = file.split('/')[-1]
        date = name.split('.')[0].split('_')[-1].replace('-', '')
        # print(date)
        return date in dates
    
    if ttype:
        files = list(filter(type_filter, files))

    # print(files)
    if dates:
        files = list(filter(date_filter, files))

    return files

# ETL / ELT - Extract Load Transform - data engineering and scrapping 
```
